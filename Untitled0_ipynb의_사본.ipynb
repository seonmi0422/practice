```python
# Reading a txt file
import pandas as pd

url = "https://raw.githubusercontent.com/seonmi0422/my_data/main/students.txt"
df = pd.read_csv(url, delimiter="\t", header=0, names=['Name', 'Age', 'Grade'])  # specify tab as delimiter and column names

# Access the columns using their actual names
df["Name"]    # Now accesses the correct column
df["Age"].mean()   # Calculate average age
```
